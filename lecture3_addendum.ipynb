#### 3.4 Addendum
To understand the physical meaning of exact ODEs, it is unavoidable to deal with some of the core concepts of vector calculus.  Here, we briefly and briskly review those concepts that are needed to understand exact ODEs.  First, consider 
a linear differential form

$$ L = P dx + Q dy  + R dz. $$

If the above equation is a total differential of a function, say $ \psi $, then it is called an exact differential or is integrable.  In other words, if $L$ is given as

$$ L = \frac{\partial \psi }{\partial x} dx + \frac{\partial \psi }{\partial y} dy + \frac{\partial \psi }{\partial z} dz $$

then its line integral over a curve $C$ is path independent or equivalently the integral over a closed curve is zero,

\begin{equation}\label{eq:exact_line_integral}\tag{1}
\oint _C L = 0.
\end{equation} 

Using vector notation, we can show this result as follows. For a scalar function (a potential function) $ \psi( x,y,z ) $ we define its gradient $\mathbf{F}$ as
$$ \mathbf{F} = \nabla \psi, $$
where $ \nabla $ is a del operator and defined as
$$ \nabla = \frac{\partial  }{\partial x} \mathbf{i} + \frac{\partial  }{\partial y} \mathbf{j} + \frac{\partial  }{\partial z} \mathbf{k} $$
in Cartesian coordinates. Thus we have

\begin{equation}\label{eq:vector_field}\tag{2}
\mathbf{F} = \frac{\partial \psi }{\partial x} \mathbf{i} + \frac{\partial \psi }{\partial y} \mathbf{j} + \frac{\partial \psi }{\partial z} \mathbf{k} = \left( \frac{\partial \psi }{\partial x}, \frac{\partial \psi }{\partial y}, \frac{\partial \psi }{\partial z} \right) = (F _1, F _2, F _3 ).
\end{equation} 

Then \eqref{eq:exact_line_integral} becomes the circulation of $ \mathbf{F} $ around a closed curve $C$ becomes $0$, i.e.,

\begin{equation}\label{eq:zero_circulation}\tag{3}
\oint _C \mathbf{F} \cdot \hat{\mathbf{t}} \, ds = 0,
\end{equation} 

since
$$ \int _C \mathbf{F} \cdot \hat{\mathbf{t}} \, ds = \int _C \frac{ d \psi }{ ds } ds = \psi (x _2, y _2, z _2 ) - \psi (x _1, y _2, z _2 ), $$
which implies the path independence. (If an integration over a path $C$ depends only on the end points, it is path independent and becomes zero when evaluated over a closed curve.  For more detail, refer to [Path Indenpendence](#addendum_line_integral).)  The first equality in the above equation comes from the following inner product,

$$ \mathbf{F} \cdot \hat{\mathbf{t}} = \frac{\partial \psi }{\partial x} \frac{ dx }{ ds } + \frac{\partial \psi }{\partial y} \frac{ dy }{ ds } + \frac{\partial \psi }{\partial z} \frac{ dz }{ ds } = \frac{ d \psi }{ ds }, $$
where, $\hat{\mathbf{t}} = \left( \frac{ dx }{ ds }, \frac{ dy }{ ds }, \frac{ dz }{ ds } \right) $ is a unit tangent vector along the curve $C$.

The zero circulation \eqref{eq:zero_circulation} implies curl $\mathbf{F}  = \nabla \times \mathbf{F} = 0 $ (see [Stoke](#stoke)), thus we have

\begin{equation}
\left( \frac{\partial F _2}{\partial z} - \frac{\partial F _3}{\partial y} \right) \mathbf{i} 
+ \left( \frac{\partial F _1}{\partial x} - \frac{\partial F _3}{\partial z} \right) \mathbf{j}
+ \left( \frac{\partial F _1}{\partial y} - \frac{\partial F _2}{\partial x} \right) \mathbf{k}
= 0,
\end{equation}
which implies 
\begin{equation}\label{eq:curl_free_check}\tag{4}
\frac{\partial F _2}{\partial z} = \frac{\partial F _3}{\partial y}, \quad  \frac{\partial F _1}{\partial x} = \frac{\partial F _3}{\partial z}, \quad \frac{\partial F _1}{\partial y} = \frac{\partial F _2}{\partial x}.
\end{equation} 

To understand these concepts graphically, let's assume $ F _3 = 0 $ and $ \psi(x,y) = x ^2 + 2 y ^2 $.  

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, y):
    return x**2 + 2*y**2

x = np.linspace(-5, 5, 40)
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

contours = plt.contour(X, Y, Z, 8, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[-5, 5, -5, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

**Path independence.**<a id='addendum_line_integral'></a>
Let $ \int _C \mathbf{F} \cdot \hat{\mathbf{t}} \, ds $ be an path independent integral.  Then we have

$$ \int _{C _1 } \mathbf{F} \cdot \hat{\mathbf{t}} \, ds = \int _{C _2 } \mathbf{F} \cdot \hat{\mathbf{t}} \, ds, $$

where $ C _1 $ and $ C _2 $ are two paths that connect from $\mathbf{x}_1$ to  $\mathbf{x}_2$.  If we reverse the direction of the second integration from $\mathbf{x}_2$ to  $\mathbf{x}_1$ (denote this as $ - C _2 $), we get

$$ \int _{C _1 } \mathbf{F} \cdot \hat{\mathbf{t}} \, ds = - \int _{-C _2 } \mathbf{F} \cdot \hat{\mathbf{t}} \, ds, $$
from which, we get
\begin{equation} 
\int _{C _1 } \mathbf{F} \cdot \hat{\mathbf{t}} \, ds - \int _{-C _2 } \mathbf{F} \cdot \hat{\mathbf{t}} \, ds = \int _{C _1 - C _2} \mathbf{F} \cdot \hat{\mathbf{t}} \, ds = 0.
\end{equation} 
**Divergence**
\begin{equation} 
\textrm{div } \mathbf{F} = \nabla \cdot V = \lim_{ \Delta V \to 0 } \frac{ 1 }{ \Delta V } \int \!\!\! \int _S  \mathbf{F} \cdot \mathbf{n} \, dS
\end{equation} 

**Divergence theorem**
\begin{equation} 
\int \!\!\! \int \!\!\! \int _V \nabla \cdot \mathbf{F} \, dV = \int \!\!\! \int _S  \mathbf{F} \cdot \mathbf{n} \, dS
\end{equation} 

**Curl**<a id='addendum_curl'></a>
\begin{equation} 
\mathbf{n} \cdot \nabla \times \mathbf{F} = \lim_{ \Delta S \to 0 } \frac{ 1 }{ \Delta S } \oint _C \mathbf{F} \cdot \mathbf{t} \, ds 
\end{equation} 

**Stoke's theorem**
\begin{equation} 
\int \!\!\! \int _S \mathbf{n} \cdot \nabla \times \mathbf{F}\,  dS = \oint _C \mathbf{F} \cdot \mathbf{t} \, ds 
\end{equation}
